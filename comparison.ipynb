Function for comparing strings and find the number of different chars.


```python
def count_different_chars(str1, str2):
    # Get the minimum length of the two strings
    min_len = min(len(str1), len(str2))

    # Count differences in the overlapping portion
    differences = sum(c1 != c2 for c1, c2 in zip(str1[:min_len], str2[:min_len]))

    # Add the absolute difference in lengths as additional differences
    differences += abs(len(str1) - len(str2))

    return differences

# Example usage:
string1 = "specises"
string2 = "species"

result = count_different_chars(string1, string2)
print(f"Number of different characters: {result}")
```

Script resulting in a new file with comparison results.

The file used has to be in specific format: one excel file containing only one sheet with headers.
(The comparison starts from the 3rd column)
We used the script with the resulted file from the "zooniverse_to_human" script.

```python
# a comparison is made per 3 rows for each column.If there are empty cells:
# if it is 2 then they remain empty and the 3rd cell keeps its value.
# if it is one then it remains empty and the other two are compared normally
# if there are no spaces:
# per 3 are also compared: if at least two are the same then we put "yes" in their positions and "no" in the position of the third.
# if less than 2 are the same, then a comparison of each 2 is made and we find the number of different chars between them, which is written as follows:
# the first cell out of 3 shows the number from the comparison of the first with the second, in the second the number from the comparison of the second and third and in the third the number from the comparison of the first and third

import pandas as pd
from openpyxl import load_workbook
from string import ascii_letters
from openpyxl.utils import get_column_letter

# Load the Excel file into a pandas DataFrame
df = pd.read_excel('pola_compare.xlsx', sheet_name=0)

# Set the number of rows to compare at a time (in this case, 3)
rows_to_compare = 3

# Create a new Excel file with the same structure
new_wb = load_workbook('pola_compare.xlsx')
new_ws = new_wb.create_sheet('New Sheet')

# Copy headers and first two columns
new_ws.append(df.columns.tolist())
for row in df.iloc[:, :2].values.tolist():
    new_ws.append(row)

# Iterate over all columns in the DataFrame (starting from the 3rd column)
for col_num, column_name in enumerate(df.columns[2:], start=3):
    # Iterate over all rows in groups of 'rows_to_compare'
    for i in range(0, len(df) - rows_to_compare + 1, rows_to_compare):
        row1 = df.iloc[i]
        row2 = df.iloc[i + 1]
        row3 = df.iloc[i + 2]

        # Normalize the strings by removing special characters and spaces
        str1 = ''.join(c for c in str(row1[column_name]) if c.isalnum() or c == '.')
        str2 = ''.join(c for c in str(row2[column_name]) if c.isalnum() or c == '.')
        str3 = ''.join(c for c in str(row3[column_name]) if c.isalnum() or c == '.')

        print(str2)
        # Compare the normalized strings
        # First look for empty strings
        comparison_result = ['no', 'no', 'no']
        if str1 == "nan":
            if str2 == "nan":
                comparison_result[0] = ""
                comparison_result[1] = ""
                comparison_result[2] = str3
            elif str3 == "nan":
                comparison_result[0] = ""
                comparison_result[1] = str2
                comparison_result[2] = ""
            else:
                # Only str1 is empty
                comparison_result[0] = ""
                if str2 == str3:
                    comparison_result[1] = 'yes'
                    comparison_result[2] = 'yes'
        elif str2 == "nan":
            if str3 == "nan":
                comparison_result[0] = str1
                comparison_result[1] = ""
                comparison_result[2] = ""
            else:
                # Only str2 is empty
                comparison_result[1] = ""
                if str1 == str3:
                    comparison_result[0] = 'yes'
                    comparison_result[2] = 'yes'
        elif str3 == "nan":
            # Only str3 is empty
            comparison_result[2] = ""
            if str1 == str2:
                comparison_result[0] = 'yes'
                comparison_result[1] = 'yes'
        else:
            if str1 == str2:
                comparison_result[0] = 'yes'
                comparison_result[1] = 'yes'
                if str1 == str3:
                    comparison_result[2] = 'yes'
            elif str1 == str3:
                comparison_result[0] = 'yes'
                comparison_result[2] = 'yes'
            elif str2 == str3:
                comparison_result[1] = 'yes'
                comparison_result[2] = 'yes'

            if comparison_result.count("yes") < 2:
                a = count_different_chars(str1, str2)
                c = count_different_chars(str1, str3)
                b = count_different_chars(str2, str3)
                comparison_result = [a, b, c]

        # Write the new values to the new sheet using the valid column name and correct row
        for j, value in enumerate(comparison_result):
            new_ws[get_column_letter(col_num) + str(i + 2 + j)] = value

# Save the changes to the existing Excel file
new_wb.save('pola_compare_res.xlsx')
```
